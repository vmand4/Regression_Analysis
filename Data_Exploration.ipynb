# Outlier Analysis

In this analysis, I would like to look at values that are at the extreme ends or highly unusual in each variable. These values are also called as outliers. One of the challenge in this data is the lack of information related to: what this data  belongs to or what each variable and target represent. As most outlier detection and analysis methods needs knowledge about the data, I just want to see the percentage of data in each variable that is > 3 standard deviations (99.7%) from mean. I used this criteria as the variables are following normal distribution.

For this purpose, I am splitting data into train test as its a better ides to perform outlier analysis on training data and then adapt the detection method to test data. Once this is done, I calculate Z scores for each variable and then calulcate the percentage of outliers based on this univariate analysis. 

I will NOT removing any outliers as I dont have knowledge about the data to remove outliers. This is analysis is just for exploration purposes.

Create a copy of data for outlier detection

```python
Main_Data_OD = Main_Data.copy()
Main_Data_OD.head()
```

# Dummy code discrete numerical variable (var_10)

As variable 10 is discrete numerical value, I converted this into dummy variables so that there wont be any natual ordering assumption that impacts regression analysis. If this variable is ordinal, then I would have gone with same values in var_10 but as I dont have any information I would treat it as a regular discrete variable and dummyh code it.

```python
Main_Data_OD = pd.get_dummies(Main_Data_OD, columns = ['var_10'],prefix = 'var_10')
Main_Data_OD.head()
```

# Univariate Outlier analysis on train data.

```python
#Seperate Predictors and Target
Main_Data_OD_X = Main_Data_OD.iloc[:,1:]
Main_Data_OD_Y = pd.DataFrame(Main_Data_OD.iloc[:,0])

#Split Data into Train Test (90:10)
Main_Data_OD_X_Train, Main_Data_OD_X_Test, Main_Data_OD_Y_Train, Main_Data_OD_Y_Test = train_test_split(Main_Data_OD_X,Main_Data_OD_Y, test_size = 0.1, random_state = 1, shuffle = False)
```

Scale data before outlier analysis as uneven scales in variables impact analysis

```python
#Scaling data
Scaler_X = StandardScaler(); Scaler_Y = StandardScaler()

# Fit scaler on Train data
Main_Data_OD_X_Train = pd.DataFrame(Scaler_X.fit_transform(Main_Data_OD_X_Train.iloc[:,:9]), columns = Main_Data_OD_X.iloc[:,:9].columns).join(Main_Data_OD_X_Train.iloc[:,9:]) 
Main_Data_OD_Y_Train = pd.DataFrame(Scaler_Y.fit_transform(Main_Data_OD_Y_Train), columns = Main_Data_OD_Y.columns)

#Transform Test data based on train data scaler parameters

Main_Data_OD_X_Test = pd.DataFrame(Scaler_X.transform(Main_Data_OD_X_Test.iloc[:,:9]), columns = Main_Data_OD_X.iloc[:,:9].columns).join(Main_Data_OD_X_Train.iloc[:,9:])
Main_Data_OD_Y_Test = pd.DataFrame(Scaler_Y.transform(Main_Data_OD_Y_Test), columns = Main_Data_OD_Y.columns)
```

Based on univariate outlier analysis, we observe that there are very less extreme samples in each data set.

```python
# Calculating percentage of values that are > 3 standard deviations from mean
OD_Percent = []
OD_Count = []
```

```python
# For loop to calculate percentage of extreme values in each variable in a pandas frame
for column in Main_Data_OD_X_Train:
    # Convert all Z scores in column to absolute values
    Temp_Data = Main_Data_OD_X_Train[column].abs()
    # Calculate Count of values >3
    C_SD_Col = (Temp_Data.where(Temp_Data>3).count())
    
    # Calculate Percentage
    P_SD_Col = ((Temp_Data.where(Temp_Data>3).count()/Temp_Data.count())*100)
    
    # Append them to list
    OD_Percent.append(P_SD_Col)
    OD_Count.append(C_SD_Col)

# Creating a dataframe to display univariate outliers in each data
OD_Percent_Count = ((pd.DataFrame(OD_Percent, index = Main_Data_OD_X_Train.columns).transpose()).
                    append(pd.DataFrame(OD_Count, index = Main_Data_OD_X_Train.columns).transpose())).set_index(pd.Series(['Percent_OD', 'Count_OD']))
                
print('Percentage & Count of values > 3 standard deviations in each variable')
OD_Percent_Count
```
